```python
# Downloading the library packages
import numpy as np
import pandas as pd
from PIL import Image
import cv2
import matplotlib.image as img
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Flatten, Conv2D, MaxPooling2D, AveragePooling2D
from tensorflow.keras import utils
from sklearn.metrics import confusion_matrix, classification_report, plot_confusion_matrix

import matplotlib.pyplot as plt
%matplotlib inline
from tensorflow import keras
from tensorflow.keras.regularizers import l1, l2
```

#### Steps
 1. Preparing the train data 
 2. Preparing the test data
 3. Creating the model
 4. Fitting and evaluating the model
 5. Predicting for test purposes
 6. Saving the model
 7. Loading the model and predicting (test purposes)


```python
# In this cell I am creating the training data. (X_train and y_train)

# A target list that will contain target values
target_list = []

# A data list that will contain the image data. We will convert it to numpy array
data_list = []

# There are 44 photos in train folder. I want to randomly select them
ph_numbers = np.random.choice(np.linspace(1,44,44), 44, replace=False)

# In order to make sure all photos are in the same size, I defined the dimensions.
width = int(780)
height = int(1196)
dimensions = (width, height)

for i in ph_numbers:
    
    # reading each photo from the file. Also converting to float data type
    string = img.imread("train/"+str(int(i))+".jpeg").astype('float32')
    
    # squeshing the photo. (normalizing)
    string /= 255
    
    #resizing the photo using cv2
    string = cv2.resize(string, dimensions, interpolation = cv2.INTER_AREA) 
    
    # appending in a list where we keep all the photo data
    data_list.append(string)
    
    # if the photo names are in the below list, we know that they are "not liked", so we add 0 for them
    if i in [16, 3, 24, 37, 10, 26, 38, 11, 27, 39, 29, 40, 14, 41, 18, 34, 42, 22, 23, 36, 43 ,44]:
        target_list.append(0)
    # if the photo names are not in the above list, they are liked.
    else:
        target_list.append(1)
        

# converting the target list into numpy array
target_list = np.array(target_list)

# converting the image data list to a numpy array
X_train = np.array(data_list)

y_train = target_list
```

```python
#loading the test data
t1 = img.imread('test/t13.jpeg').astype('float32') / 255
t2 = img.imread('test/t14.jpeg').astype('float32') / 255
t3 = img.imread('test/t15.jpeg').astype('float32') / 255
t4 = img.imread('test/t10.jpeg').astype('float32') / 255
t5 = img.imread('test/t11.jpeg').astype('float32') / 255
t6 = img.imread('test/t12.jpeg').astype('float32') / 255


#resizing the photos
t1 = cv2.resize(t1, dimensions, interpolation = cv2.INTER_AREA) 
t2 = cv2.resize(t2, dimensions, interpolation = cv2.INTER_AREA) 
t3 = cv2.resize(t3, dimensions, interpolation = cv2.INTER_AREA) 
t4 = cv2.resize(t4, dimensions, interpolation = cv2.INTER_AREA) 
t5 = cv2.resize(t5, dimensions, interpolation = cv2.INTER_AREA) 
t6 = cv2.resize(t6, dimensions, interpolation = cv2.INTER_AREA) 


# creating a numpy array of X variables
X_test = np.array([t1 , t2 , t3 , t4 , t5 , t6])


# creating target values, based on the numpy array sequence
y_test = np.array([0,0,0,1,0,0]) 

# converting target test array to categorical array
y_test = utils.to_categorical(y_test)
```

```python
# Instantiate a CNN.
cnn_model = Sequential()

# Add a convolutional layer with 8 filters
cnn_model.add(Conv2D(
                    filters=8,
                    kernel_size=(3,3),
                    activation='relu',
                    input_shape=(1196, 780, 3)) 
             )

# pooling using MaxPool2D
cnn_model.add(MaxPooling2D(pool_size=(3,3)))


# Add a convolutional layer with 16 filters
cnn_model.add(Conv2D(
                filters=16,
                kernel_size=(3, 3),
                activation='relu',
                input_shape=(1196, 780, 3))
              )
# pooling using MaxPool2D (1 for after each convolutional layer)
cnn_model.add(MaxPooling2D(pool_size=(3,3)))


# Add a convolutional layer with 32 filters
cnn_model.add(Conv2D(
                filters=32,
                kernel_size=(3, 3),
                activation='relu',
                input_shape=(1196, 780, 3))
              )
# pooling using MaxPool2D (1 for after each convolutional layer)
cnn_model.add(MaxPooling2D(pool_size=(3,3)))

# Add a convolutional layer with 64 filters
cnn_model.add(Conv2D(
                filters=64,
                kernel_size=(3, 3),
                activation='relu',
                input_shape=(1196, 780, 3))
              )
# pooling using MaxPool2D (1 for after each convolutional layer)
cnn_model.add(MaxPooling2D(pool_size=(3,3)))

# Flattening the data for dense layers
cnn_model.add(Flatten())

# Add a densely-connected layer with 32 neurons.
cnn_model.add(Dense(32, activation='relu'))
# Using dropout regularizer
cnn_model.add(Dropout(0.3)) 

# Add a densely-connected layer with 32 neurons.
cnn_model.add(Dense(32, activation='relu'))
# Using dropout regularizer
cnn_model.add(Dropout(0.3)) 

# Add a densely-connected layer with 32 neurons.
cnn_model.add(Dense(32, activation='relu'))
# Using dropout regularizer
cnn_model.add(Dropout(0.5)) 

# Add a final layer with 2 neurons and Ridge regularization
cnn_model.add(Dense(2, activation='softmax', 
                    kernel_regularizer=l2(0.0001)
                   ))


# Compile model
cnn_model.compile( 
                    loss='categorical_crossentropy',
                  optimizer='adam',
                  metrics=['accuracy'])
```

```python
# Looking at the summary of each output shape
cnn_model.summary()
```

```python
# Fit model on training data and testing
history = cnn_model.fit(X_train,
                        y_train,
                        batch_size=2,
                        validation_data=(X_test, y_test),
                       epochs=10 )
```

```python
# Ploting the accuracy
plt.figure(figsize=(9, 5))

plt.plot(history.history["accuracy"], label="Train Accuracy", c="green")
plt.plot(history.history["val_accuracy"], label="Test Accuracy", c="blue")
plt.legend()
plt.title("accuracy rate for each epoch", );
```

```python
# Ploting the loss

plt.figure(figsize=(9, 5))

plt.plot(history.history["loss"], label="Train Loss", c="red")
plt.plot(history.history["val_loss"], label="Test Loss", c="blue")
plt.legend()
plt.title("loss rate for each epoch", );
```

```python
# evaluatin the model one more time with test data
cnn_model.evaluate(X_test,y_test)
```
